```python
# Now, import spaCy and perform named entity recognition
import spacy
!{sys.executable} -m spacy download xx_ent_wiki_sm

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

# Define the text to be analyzed
text = (
    "On September 15, 2023, Dr. John Smith, a leading researcher at GeeksforGeeks, "
    "presented a groundbreaking paper at the International Conference on Online Learning held in Mumbai, India. "
    "The event was attended by various experts from around the globe, including representatives from organizations such as "
    "Tech Innovations Inc. and Global Learning Solutions. "
    "Dr. Smith emphasized the importance of online education in today's digital age, citing numerous studies and statistics. "
    "Many attendees expressed their excitement about the future of e-learning and the potential it has to transform education. "
    "The conference will continue until September 20, 2023, showcasing various workshops and keynote speeches."
)

# Process the text with spaCy
doc = nlp(text)

# Print the entities found in the text
for ent in doc.ents:
    print(ent.label_, '-->', ent.text)
```

```python
import sys
import spacy
from googletrans import Translator

# Load the multilingual spaCy model
nlp = spacy.load("xx_ent_wiki_sm")

# Initialize the Google Translator
translator = Translator()

# Define a Marathi text to be analyzed
marathi_text = (
    "१५ सप्टेंबर २०२३ रोजी, डॉ. जॉन स्मिथ, गीक्सफॉरगीक्स येथे एक आघाडीचा संशोधक, "
    "मुंबई, भारत येथे झालेल्या आंतरराष्ट्रीय ऑनलाइन शिक्षण परिषदेत एक क्रांतिकारी पेपर सादर केला. "
    "या कार्यक्रमात जगभरातून विविध तज्ञ उपस्थित होते, जसे की टेक इनोव्हेशन्स इंक. "
    "आणि जागतिक शिक्षण उपाय. डॉ. स्मिथने आजच्या डिजिटल युगात ऑनलाइन शिक्षणाच्या महत्त्वावर जोर दिला, "
    "अन्य अनेक अभ्यास आणि आकडेवारी उद्धृत केली. अनेक उपस्थितांनी ई-शिक्षणाच्या भविष्याबद्दल "
    "आवड व्यक्त केली आणि यामुळे शिक्षण कसे बदलू शकते याबद्दल चर्चा केली. "
    "ही परिषद २० सप्टेंबर २०२३ पर्यंत सुरू राहील, विविध कार्यशाळा आणि मुख्य भाषणांची प्रदर्शनी होईल."
)

# Translate the Marathi text to English
english_text = translator.translate(marathi_text, src='mr', dest='en').text

# Process the English text with spaCy
doc_en = nlp(english_text)

# Print the English translation
print("Translated English Text:\n", english_text)

# Print the entities found in the English text
print("\nEntities found in English Text:")
for ent in doc_en.ents:
    print(ent.label_, '-->', ent.text)
```

```python

# Sample text in Marathi
text_mr = "संयुक्त राष्ट्रांचे सरचिटणीस अंतोनियो गुटेरेस यांनी नोव्हेंबर २०२३ मध्ये भारताला भेट दिली. आपल्या भेटीदरम्यान, त्यांनी पंतप्रधान नरेंद्र मोदी यांची नवी दिल्लीत पंतप्रधान कार्यालयात भेट घेतली. या भेटीत त्यांनी हवामान बदल, जागतिक शांतता आणि सुरक्षा, तसेच शाश्वत विकास उद्दिष्टे यासारख्या विविध मुद्द्यांवर चर्चा केली. गुटेरेस यांनी भारतीय संसदेला देखील संबोधित केले, जिथे त्यांनी जगातील तातडीच्या समस्या सोडवण्यासाठी अधिक आंतरराष्ट्रीय सहकार्याची मागणी केली. भारताच्या भेटीनंतर, गुटेरेस बांगलादेशला गेले, जिथे त्यांनी पंतप्रधान शेख हसीना यांची भेट घेतली आणि रोहिंग्या शरणार्थी संकटावर चर्चा केली."

# Process the text
doc = nlp(text_mr)

# Print the detected entities
for sentence in doc.sentences:
    for entity in sentence.ents:
        print(f"{entity.text} : {entity.type}")
```

```python


text_mr = ("संयुक्त राष्ट्र संघाचे जनरल सेक्रेटरी अंतोनियो गुटेरेस १५ नोव्हेंबर २०२३ रोजी मुंबईत आले. "
            "त्यांनी भारतीय पंतप्रधान नरेंद्र मोदी आणि महाराष्ट्राचे मुख्यमंत्री एकनाथ शिंदे यांच्याशी चर्चा केली. "
            "या भेटीत त्यांनी जलवायु बदल, जागतिक शांतता आणि विकास यांच्या विषयी विचार मांडले. "
            "गुटेरेस यांनी 'ताज महाल' या प्रसिद्ध भारतीय स्मारकाला भेट दिली आणि या ऐतिहासिक स्थळाची प्रशंसा केली. "
            "त्यानंतर, ते बांगलादेशात गेले जिथे त्यांनी पंतप्रधान शेख हसीना यांची भेट घेतली आणि रोहिंग्या शरणार्थी संकटावर चर्चा केली. "
            "या दौऱ्याच्या शेवटी, गुटेरेस यांनी 'न्यूयॉर्क टाइम्स'साठी एक मुलाखत दिली.")


# Define a mapping from entity types to Marathi
entity_type_mapping = {
    'NEP': 'व्यक्ती',       # Person
    'NEO': 'संस्था',       # Organization
    'NEL': 'स्थान',       # Location (Geopolitical Entities, etc.)
    'LOC': 'ठिकाण',       # Location (Non-GPE locations)
    'NED': 'तारीख',       # Date
    'TIME': 'वेळ',        # Time
    'MONEY': 'पैसे',      # Monetary values
    'QUANTITY': 'प्रमाण',  # Measurements
    'ORDINAL': 'क्रमांक',  # Ordinals
    'CARDINAL': 'संख्या',  # Numerals
    'EVENT': 'कार्यक्रम',  # Named events
    'PRODUCT': 'उत्पादन',  # Products
    'LAW': 'कायदा',        # Laws
    'LANGUAGE': 'भाषा',    # Languages
    'PERCENT': 'टक्के',    # Percentages
    'WORK_OF_ART': 'कलाकृती',  # Works of art
    'FAC': 'सुविधा',       # Facilities
    'NORP': 'गट',          # Nationalities, religions, and political groups
    'MISC': 'इतर',        # Miscellaneous
    'ED': 'पदवी'          # Titles/Designations
}

# Process the Marathi text
doc = nlp(text_mr)

# Print the detected entities with Marathi labels
for sentence in doc.sentences:
    for entity in sentence.ents:
        entity_type_marathi = entity_type_mapping.get(entity.type, entity.type)
        print(f"{entity.text} - {entity_type_marathi}")
```
